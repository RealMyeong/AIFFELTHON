```python
def get_attker(conv):
    try:
        sentences = conv.split('\n')
    except:
        return conv
        
    attk_only = ''
    for i in range(len(sentences)):
        try:
            talker, sen = sentences[i].split(':', maxsplit=1)
            try:
                if talker == '1': 
                    if attk_only != '': attk_only = attk_only + '\n'
                    attk_only = attk_only + sen
            except:
                continue
        except:
            continue

    return attk_only
```

```python
ch['sentence'] = ch['label_sentence'].apply(lambda x: get_attker(x))
```

```python
ch = ch[['class', 'sentence']]
```

```python
ori = ori[['class', 'sentence']]
```

```python
ori_ch = pd.concat([ori, ch], axis=0)
```
