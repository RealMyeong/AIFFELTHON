```python
import pandas as pd
import json
```

```python
train = pd.read_csv('/content/drive/MyDrive/공부/AIFFELTHON/data/train.csv')
```

```python
mask = pd.read_csv('/content/drive/MyDrive/공부/AIFFELTHON/data/train_mask.csv', encoding='utf-8')
```

```python
train['count'] = train.apply(lambda x : x['conversation'].count('\n') + 1 , axis = 1 )
```

```python
all_sentence = []
all_labels = []

for i in range(len(mask)):  
  talker1 = []
  talker2 = []
  mask = train['mask'][i]
  sen = train['conversation'][i].split('\n')
  for j in range(12):
    m = (mask[str(j + 1)][i])
    if m == 1:
      talker1.append(sen[j])
    elif m == 2:
      talker2.append(sen[j])
    elif m == None:
      continue
  reu_sen = '\n'.join(talker1)
  label = train['class'][i] + ' 위협자'

  all_sentence.append(reu_sen)
  all_labels.append(label)

  reu_sen = '\n'.join(talker2)
  label = train['class'][i] + ' 피해자'
  
  all_sentence.append(reu_sen)
  all_labels.append(label)

```

```python
#new_data = pd.DataFrame(data = [[all_sentence, all_labels]], columns = ['conversation', 'class'])
new_data = pd.DataFrame({'conversation':all_sentence, 'class':all_labels})
```
