```python
from textaugment import EDA

t = EDA()

def conv_swap(conversation):
  sentences = conversation.split('\n')
  swapped_conv = ''
  for i in range(len(sentences)):
    if i != 0: swapped_conv += '\n'
    
    if len(sentences[i].split(' ')) > 1:
      swapped_conv = swapped_conv + t.random_swap(sentences[i])
  
  return swapped_conv
```

```python
import pandas as pd

data = pd.read_csv('/content/drive/MyDrive/공부/AIFFELTHON/shared/TUNiB/train_normalization.csv')
data.head()
```

```python
aug_data['non_label_sentence'] = data['non_label_sentence'].apply(lambda x: conv_swap(x))
```
